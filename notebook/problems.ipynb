## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import random

def simulate_two_dice_rolls(simulations):
    count_even_or_greater_than_7 = 0

    for throw in range(simulations):
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        sum_results = die1 + die2

        if sum_results > 7 or sum_results % 2 == 0:
            count_even_or_greater_than_7 += 1

    probability = count_even_or_greater_than_7 / simulations
    return probability

simulations = 1000
probability = simulate_two_dice_rolls(simulations)
print(f"Probability that the total score is even or greater than 7: {probability:.4f}")

```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
simulations = 1000
count_3_white_2_red = 0
count_same_color = 0

for throw in range(simulations):
    picked_balls = random.choices(list(ball_box.values()), k=5)
    
    if picked_balls.count("White") == 3 and picked_balls.count("Red") == 2:
        count_3_white_2_red += 1
        
    if len(set(picked_balls)) == 1:
        count_same_color += 1

probability_3_white_2_red = count_3_white_2_red / simulations
probability_same_color = count_same_color / simulations

print(f"Probability of drawing 3 white and 2 red balls: {probability_3_white_2_red:.4f}")
print(f"Probability of drawing 5 balls of the same color: {probability_same_color:.4f}")
```
